```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, PowerTransformer, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from xgboost import XGBRegressor

def train_and_evaluate_disease_model(
    csv_path,
    test_size = 0.2,
    random_state = 42
):

    # 1. Caricamento e melting del dataset
    df = pd.read_csv(csv_path)
    id_vars = ["Country/Territory", "Code", "Year"]
    value_vars = [c for c in df.columns if c not in id_vars]
    df_long = df.melt(
        id_vars=id_vars,
        value_vars=value_vars,
        var_name="Disease",
        value_name="Deaths"
    )

    # 2. Calcolo della probabilitÃ  target
    df_long["TotalDeaths"] = (
        df_long.groupby(["Country/Territory", "Year"])["Deaths"]
        .transform("sum")
    )
    df_long["Prob"] = df_long["Deaths"] / df_long["TotalDeaths"]

    # 3. Definizione di X e y
    X = df_long[["Country/Territory", "Year", "Disease"]]
    y = df_long["Prob"]

    # 4. Split train/test
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )

    # 5. Preprocessing: PowerTransformer + StandardScaler per Year, OneHot per categoriche
    numeric_features = ["Year"]
    numeric_transformer = Pipeline([
        ("power", PowerTransformer(method="yeo-johnson")),
        ("scaler", StandardScaler())
    ])

    categorical_features = ["Country/Territory", "Disease"]
    categorical_transformer = OneHotEncoder(handle_unknown="ignore")

    preprocessor = ColumnTransformer([
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ])

    # 6. Costruzione della pipeline con XGBoost
    model = Pipeline([
        ("preprocessor", preprocessor),
        ("regressor", XGBRegressor(
            objective="reg:squarederror",
            random_state=random_state,
            n_estimators=100
        ))
    ])

    # 7. Fit del modello
    model.fit(X_train, y_train)

    # 8. Previsioni e metriche
    y_pred = model.predict(X_test)
    r2   = r2_score(y_test, y_pred)
    mae  = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    metrics = {
        "r2":   r2,
        "mae":  mae,
        "rmse": rmse
    }

    return model, metrics

# Esempio di utilizzo in un notebook:
model, metrics = train_and_evaluate_disease_model(r"C:\Users\fabri\Desktop\MarcoPatierno_DepositoCorsoPython\env\Giorno 21 05-05\esercitazione\data\cause_of_deaths.csv")
print(f"RÂ²:   {metrics['r2']:.4f}")
print(f"MAE:  {metrics['mae']:.4f}")
print(f"RMSE: {metrics['rmse']:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, PowerTransformer, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from xgboost import XGBRegressor

def train_and_evaluate_with_gridsearch(
    csv_path,
    test_size=0.2,
    random_state=42,
    cv=5,
    param_grid=None
):

    # Caricamento e melting
    df = pd.read_csv(csv_path)
    id_vars = ["Country/Territory", "Code", "Year"]
    value_vars = [c for c in df.columns if c not in id_vars]
    df_long = df.melt(id_vars=id_vars, value_vars=value_vars,
                      var_name="Disease", value_name="Deaths")
    
    # ProbabilitÃ  target
    df_long["TotalDeaths"] = df_long.groupby(
        ["Country/Territory", "Year"]
    )["Deaths"].transform("sum")
    df_long["Prob"] = df_long["Deaths"] / df_long["TotalDeaths"]
    
    # X e y
    X = df_long[["Country/Territory", "Year", "Disease"]]
    y = df_long["Prob"]
    
    # Split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )
    
    # Preprocessing
    numeric_features = ["Year"]
    numeric_transformer = Pipeline([
        ("power", PowerTransformer(method="yeo-johnson")),
        ("scaler", StandardScaler())
    ])
    categorical_features = ["Country/Territory", "Disease"]
    categorical_transformer = OneHotEncoder(handle_unknown="ignore")
    preprocessor = ColumnTransformer([
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ])
    
    # Pipeline base
    pipeline = Pipeline([
        ("preprocessor", preprocessor),
        ("regressor", XGBRegressor(
            objective="reg:squarederror", random_state=random_state
        ))
    ])
    
    # Param grid di default
    if param_grid is None:
        param_grid = {
            "regressor__n_estimators": [50, 100, 200],
            "regressor__max_depth": [3, 5, 7],
            "regressor__learning_rate": [0.01, 0.1, 0.2]
        }
    
    # GridSearchCV
    grid_search = GridSearchCV(
        pipeline,
        param_grid=param_grid,
        cv=cv,
        scoring="neg_mean_absolute_error",
        n_jobs=-1
    )
    grid_search.fit(X_train, y_train)
    
    # Miglior modello
    best_model = grid_search.best_estimator_
    best_params = grid_search.best_params_
    
    # Valutazione su test set
    y_pred = best_model.predict(X_test)
    metrics = {
        "r2": r2_score(y_test, y_pred),
        "mae": mean_absolute_error(y_test, y_pred),
        "rmse": np.sqrt(mean_squared_error(y_test, y_pred))
    }
    
    return best_model, metrics, best_params

best_model, metrics, best_params = train_and_evaluate_with_gridsearch(r"C:\Users\fabri\Desktop\MarcoPatierno_DepositoCorsoPython\env\Giorno 21 05-05\esercitazione\data\cause_of_deaths.csv")
print("Best params:", best_params)
print(f"RÂ²: {metrics['r2']:.4f}, MAE: {metrics['mae']:.4f}, RMSE: {metrics['rmse']:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, PowerTransformer, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from xgboost import XGBRegressor

def train_and_evaluate_with_gridsearch(
    csv_path,
    test_size=0.2,
    random_state=42,
    cv=5,
    param_grid=None
):
    # 1. Caricamento e melting
    df = pd.read_csv(csv_path)
    id_vars = ["Country/Territory", "Code", "Year"]
    value_vars = [c for c in df.columns if c not in id_vars]
    df_long = df.melt(
        id_vars=id_vars,
        value_vars=value_vars,
        var_name="Disease",
        value_name="Deaths"
    )
    
    # 2. ProbabilitÃ  target
    df_long["TotalDeaths"] = (
        df_long.groupby(["Country/Territory", "Year"])["Deaths"]
        .transform("sum")
    )
    df_long["Prob"] = df_long["Deaths"] / df_long["TotalDeaths"]
    
    # 3. Definizione di X, y e split
    X = df_long[["Country/Territory", "Year", "Disease"]]
    y = df_long["Prob"]
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )
    
    # 4. Preprocessing
    numeric_features = ["Year"]
    numeric_transformer = Pipeline([
        ("power", PowerTransformer(method="yeo-johnson")),
        ("scaler", StandardScaler())
    ])
    categorical_features = ["Country/Territory", "Disease"]
    categorical_transformer = OneHotEncoder(handle_unknown="ignore")
    preprocessor = ColumnTransformer([
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ])
    
    # 5. Pipeline + GridSearchCV
    pipeline = Pipeline([
        ("preprocessor", preprocessor),
        ("regressor", XGBRegressor(
            objective="reg:squarederror",
            random_state=random_state
        ))
    ])
    
    if param_grid is None:
        param_grid = {
            "regressor__n_estimators": [50, 100, 200],
            "regressor__max_depth": [3, 5, 7],
            "regressor__learning_rate": [0.01, 0.1, 0.2]
        }
    
    grid_search = GridSearchCV(
        pipeline,
        param_grid=param_grid,
        cv=cv,
        scoring="neg_mean_absolute_error",
        n_jobs=-1
    )
    grid_search.fit(X_train, y_train)
    
    best_model  = grid_search.best_estimator_
    best_params = grid_search.best_params_
    
    # 6. Metriche sul TRAIN set
    y_train_pred = best_model.predict(X_train)
    train_r2   = r2_score(y_train, y_train_pred)
    train_mae  = mean_absolute_error(y_train, y_train_pred)
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    print("=== Metriche sul TRAIN set ===")
    print(f"RÂ² (train):   {train_r2:.4f}")
    print(f"MAE (train):  {train_mae:.4f}")
    print(f"RMSE (train): {train_rmse:.4f}\n")
    
    # 7. Metriche sul TEST set
    y_test_pred = best_model.predict(X_test)
    test_metrics = {
        "r2":   r2_score(y_test, y_test_pred),
        "mae":  mean_absolute_error(y_test, y_test_pred),
        "rmse": np.sqrt(mean_squared_error(y_test, y_test_pred))
    }
    print("=== Metriche sul TEST set ===")
    print(f"RÂ² (test):   {test_metrics['r2']:.4f}")
    print(f"MAE (test):  {test_metrics['mae']:.4f}")
    print(f"RMSE (test): {test_metrics['rmse']:.4f}")
    
    # 8. Lista delle malattie
    disease_list = df_long["Disease"].unique().tolist()
    
    return best_model, test_metrics, best_params, disease_list

def predict_top_disease(model, disease_list, country, year):
    X_new = pd.DataFrame([{
        "Country/Territory": country,
        "Year": year,
        "Disease": d
    } for d in disease_list])
    probs  = model.predict(X_new)
    idx_max = np.argmax(probs)
    return disease_list[idx_max], probs[idx_max]

# Esempio di utilizzo:
best_model, test_metrics, best_params, diseases = train_and_evaluate_with_gridsearch(
    r"C:\Users\fabri\Desktop\MarcoPatierno_DepositoCorsoPython\env\Giorno 21 05-05\esercitazione\data\cause_of_deaths.csv"
)
disease, prob = predict_top_disease(best_model, diseases, "Italy", 2025)
print(f"La malattia piÃ¹ probabile Ã¨ {disease} ({prob:.2%})")
```
